## 1) Check if a list is sorted

Check if a given list of **Strings** is sorted in ascending order lexicographically. 
Also, report the time complexity of your algorithm.

Below is an example signature for your approach:
```python
def is_sorted(list):
    return False
```

Example input/outputs:
``` toml
- Input: ["apple", "Banana", "cherry"]      ➡️ Output: True
- Input: [12, 'car', 'deal']                ➡️ Output: False
- Input: ['purple', 'menu', 'abracadabra']  ➡️ Output: False
- Input: [12]                               ➡️ Output: False
- Input: []                                 ➡️ Output: False
- Input: 12                                 ➡️ Output: False
```

### Solution

````python
def is_sorted(list):
    try:
        if len(list) > 1:
            # Compare each item with the previous one,
            # if we find any item that is smaller than the previous one, the list is not sorted
            # Time complexity of the algorithm: O(n), where n is the length of the list
            for i in range(1, len(list)):
                # We are targeting only strings
                if isinstance(list[i], str) and isinstance(list[i - 1], str):
                    if list[i].lower() < list[i - 1].lower():
                        return False
                else:
                    return False
            return True

        # If not enough items to compare,
        # we then check if the single item is a string, a single string is considered sorted
        elif len(list) == 1 and isinstance(list[0], str):
            return True
        else:
            return False

    # if the input is not a list, or the list contains incomparable types (e.g., int and str), a TypeError will be raised
    except TypeError:
        return False
````

````python
list1 = ["apple", "Banana", "cherry"]
print("Is", list1, "sorted?", is_sorted(list1))  # True in lexicographical order

list2 = [1, 3, 2, 4, 5]
print("Is", list2, "sorted?", is_sorted(list2))  # False

list3 = []
print(
    "Is", list3, "sorted?", is_sorted(list3)
)  # False, because there are no items to compare

list4 = [5]
print("Is", list4, "sorted?", is_sorted(list4))  # False

list5 = [1, "two", 3.0]
print(
    "Is", list5, "sorted?", is_sorted(list5)
)  # False, because int, str, and float cannot be compared with each other

print(
    "Is 'not a list' sorted?", is_sorted("not a list")
)  # False, because the input is not a list

list6 = [12, "car", "deal"]
print("Is", list6, "sorted?", is_sorted(list6))  # False

list7 = ["purple", "menu", "abracadabra"]
print("Is", list7, "sorted?", is_sorted(list7))  # False

list8 = ["abracadabra", "menu", "purple"]
print("Is", list8, "sorted?", is_sorted(list8))  # True
````

## 2) Find the sum of three median elements

Print the sum of the three median elements of a sorted list in descending order.
- Your given list may or may not be sorted. You can solve this task with or without sorting the list 👀 that is up to you.
- If the list has an odd number of elements, select the three elements centered around the middle index.
- If the list has an even number of elements, select the three elements closest to the end of the list among the central elements. In other words, take the middle pair and include the next element toward the end.
- If the list has fewer than three elements or contains non numeric values, print an appropriate error message.
- Report the time and space complexity of your approach.

Example input/outputs:
``` toml
Input:  [1, 5, 3, 9, 7] - when Sorted: [9, 7, 5, 3, 1]
Middle elements: [7, 5, 3]
Output: 15

Input:  [4, 8, 2, 6] - when Sorted: [8, 6, 4, 2]
Middle elements (closer to end): [6, 4, 2]
Output: 12
```

### Solution

````python
import heapq

"""
Here the time complexity of the algorithm is O(n log k), 
    where n is the length of the list and k is the number of largest items we need to find.

The space complexity is O(k) for storing the k largest items.

** However, since k is approximating to n, 
    we can overall say that the time complexity is O(n log n) and the space complexity is O(n).
"""


def median_sum(list):
    if len(list) < 3 or not all(isinstance(x, (int, float)) for x in list):
        return "Invalid input"
    n = len(list)
    # It is indeed a k-largest problem, we need to find the 3 largest items around the median
    k = n // 2 + 2
    top = heapq.nlargest(k, list)
    mid = n // 2
    return sum(top[mid - 1 : mid + 2])
````

````python
print(median_sum([1, 2, 3, 4, 5]))  # 9 = 4 + 3 + 2
print(median_sum([5, 3, 1, 4]))  # 12 = 4 + 3 + 1
print(median_sum([1, 2]))  # "Invalid input" (not enough items)
print(median_sum([1, "two", 3]))  # "Invalid input" (not all items are numbers)
print(median_sum([1, 1099, -1]))  # 1099 = 1099 + 1 + (-1)
````
