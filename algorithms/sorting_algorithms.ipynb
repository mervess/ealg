# Sorting Algorithms

Sorting is a fundamental operation that helps us understand how algorithm efficiency scales with data size and introduces key concepts such as recursion and trade offs between time and memory.

In data science, sorting is widely used for tasks like ranking, organising, and selecting top elements, so understanding it helps you make more efficient and informed decisions when working with data.

Below are selected, touchstone sorting algorithms that offer different approaches to sorting.


| Algorithm  | Description                                                                 | Best Case  | Average Case | Worst Case | Space Complexity | Notes                                      | Intuition                                     |
|------------|-----------------------------------------------------------------------------|------------|--------------|------------|------------------|--------------------------------------------|-----------------------------------------------|
| Merge Sort | Divide and conquer algorithm that recursively splits the array and merges sorted halves  | $O(n log n)$ | $O(n log n)$   | $O(n log n)$ | $O(n)$             | Stable, requires extra memory            | Split → sort halves → merge                  |
| Heap Sort  | Builds a heap and repeatedly extracts the maximum or minimum element                     | $O(n log n)$ | $O(n log n)$   | $O(n log n)$ | $O(1)$             | Unstable, in place                       | Build heap → repeatedly extract root         |
| Quick Sort | Partition based algorithm using a pivot to sort elements around it                       | $O(n log n)$ | $O(n log n)$   | $O(n^2)$     | $O(log n)$         | Unstable, very fast in practice, pivot sensitive     | Pick pivot → split into two sides → recurse  |

## Sorting in Python

Python has `list.sort()` for an in-place sort and `sorted()` function that creates and returns a new sorted list.

Python uses [Timsort](https://en.wikipedia.org/wiki/Timsort) algorithm _(mixture of merge sort and insertion sort)_ for sorting.

| Algorithm | Description                                              | Best Case   | Average Case     | Worst Case       | Space Complexity | Notes                                      | Intuition                                      |
|-----------|----------------------------------------------------------|-------------|------------------|------------------|------------------|--------------------------------------------|-----------------------------------------------|
| Timsort   | Hybrid of merge sort and insertion sort, used in Python  | $O(n)$      | $O(n \log n)$    | $O(n \log n)$    | $O(n)$           | Stable, exploits existing order in data    | Detect sorted runs → extend → merge efficiently |

### Examples

#### `list.sort()`

```python
list = [1, 99, -101, 35, 3.4]
print("list:", list)
list.sort()
print("sorted list:", list)
```

#### `sorted()`

```python
mixed_list = ["1", "two", "3.4", "forty", "5"]
print("mixed_list:", mixed_list)
print("sorted mixed_list:", sorted(mixed_list))
print("mixed_list:", mixed_list)

# The original list is unchanged, sorted() returns a new sorted list
# Beware that:
# 1) The items should be of the same type, otherwise Python will raise a TypeError.
# 2) Sorting a list of strings will sort them lexicographically.
```

#### `sorted()` with `key` and `reverse`

```python
from operator import itemgetter

students = [
    {"name": "Alice", "grade": 85},
    {"name": "Bob", "grade": 92},
    {"name": "Charlie", "grade": 78},
]

# Sort by grade, in descending order
sorted_students = sorted(students, key=itemgetter("grade"), reverse=True)
print("Students sorted by grade:", *sorted_students, sep="\n")
```

#### Getting min-_k_ and top-_k_ items via `heapq`

```python
import heapq

data = [10, 3, 5, 8, 2]
print("Two smallest:", heapq.nsmallest(2, data))  # [2, 3]
print("Three largest:", heapq.nlargest(3, data))  # [10, 8, 5]
```
