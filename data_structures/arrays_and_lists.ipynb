# Arrays, Array Lists, and Linked Lists in Python 🐍


## 1) Arrays

The built in `array` module stores elements of the same type.

- Type enforcement.
- Elements are stored in contiguous memory.
- Fixed size in many languages.
- Index access is constant time.
- Efficient cache behavior.

```python
from array import array

arr = array("i", [1, 2, 3, 4])
print(arr)

arr.append(5)
print("After append:", arr)

print("**Type of the array:", type(arr))

try:
    arr.append(3.14)
except TypeError as e:
    print("TypeError:", e)
```

## 2) Array Lists

Python `list` behaves like an array-list. This is an array structure under the hood, not a chain of nodes.

- Dynamic array that resizes automatically.
- Internally backed by a contiguous memory block.
- Supports fast index access.
- `append` is amortized constant time → amortized O(1). Can you guess why?

For numerical computing, *NumPy* arrays are common. 
*You'll see an example of that during the session "Time&Memory Complexity".*

```python
# Python list example - I

a_list = [10, "20", 30.56, -40]
print("Array-list:", a_list)
print("First element:", a_list[0])
print("Third element:", a_list[2])

# Appending a new element at the end of the list
a_list.append(50)
print("Array-list after append:", a_list)

# Let's see the types of the elements in the list + the type of the list itself
print(
    "Type of first element:", type(a_list[0])
)  # This will show the type of the first element, which is int
print(
    "Type of second element:", type(a_list[1])
)  # This will show the type of the second element, which is str
print(
    "Type of third element:", type(a_list[2])
)  # This will show the type of the third element, which is float

print(
    "**Type of the list itself:", type(a_list)
)  # This will show the type of the list itself, which is <class 'list'>
```

```python
# Python list example - II

a_list_v2 = list(range(10))
print("Initial:", a_list_v2)

a_list_v2.insert(0, 999)
print("After insert at start:", a_list_v2)

a_list_v2.pop(0)
print("After 'pop' at start:", a_list_v2)

a_list_v2.append(111)
print("After append:", a_list_v2)

print(
    "**Type of the list itself:", type(a_list_v2)
)  # This will show the type of the list itself, which is <class 'list'>
```

## 3) Linked Lists

Python’s `collections.deque` -which was designed for queues and stacks- is implemented as a doubly linked structure of blocks. It is the closest thing in the standard library to a linked list abstraction.

Linked lists do not store elements contiguously.

Each node contains:
- value
- reference to next node

Tradeoff:
- No shifting cost.
- No constant time indexing.
- Higher memory overhead due to references.


```python
from collections import deque

linked_list = deque([1, 2, 3])
linked_list.appendleft(0)
print("After appendleft:", linked_list)

print(
    "**Type of the list itself:", type(linked_list)
)  # This will show the type of the list itself, which is <class 'deque'>
```

### **Example:** Implementing your own Linked List in Python

```python
class Node:
    def __init__(self, value, next_node=None):
        self.value = value
        self.next = next_node

    def __repr__(self):
        return f"Node({self.value})"


class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        # Creating a new node with the given value
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            return

        # Linking the new node to the end of the list
        current_node = self.head
        while current_node.next is not None:
            current_node = current_node.next
        current_node.next = new_node

    def find(self, value):
        current_node = self.head
        while current_node is not None:
            if current_node.value == value:
                return current_node
            current_node = current_node.next
        return None

    def to_list(self):
        result = []
        current_node = self.head
        while current_node is not None:
            result.append(current_node.value)
            current_node = current_node.next
        return result

    def __repr__(self):
        return f"LinkedList({self.to_list()})"
```

The differences between these data structures are most noticeable in their memory usage.
Below, we will examine that.

```python
# Installing pympler to measure memory usage of the data structures
!pip install pympler
from pympler import asizeof as sizer
# Note that pympler measures the memory usage of the entire object (in-memory Python object graph).

#---------------------------------------------- Array ---------------------------------------
from array import array

arr = array("i", [10, 20, 30])
print("Size of array:", sizer.asizeof(arr), "bytes")

#---------------------------------------- Array List ----------------------------------------
a_list = [10, 20, 30]
print("Size of array list:", sizer.asizeof(a_list), "bytes")

#---------------------------------------- Linked List ---------------------------------------
linked_list = LinkedList()
linked_list.append(10)
linked_list.append(20)
linked_list.append(30)

print("Size of custom linked list:", sizer.asizeof(linked_list), "bytes")

# Comparing with deque from collections
from collections import deque

dlinked_list = deque([10, 20, 30])
print("Size of deque:", sizer.asizeof(dlinked_list), "bytes")
```
