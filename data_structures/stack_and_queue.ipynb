# Stack

Stack is a *last-in-first-out* (LIFO) data structure and is realised via "lists" in python.

Resource: [python docs](https://docs.python.org/3.13/tutorial/datastructures.html#using-lists-as-stacks)

![Stack Image Display](stack.svg)

````python
stack = ["apple", "banana", "cherry"]
# Here we call "append" as it uses a list internally.
# In pseudo code, it would be called "push" for stacks.
stack.append("date")
stack.append("elderberry")
stack
````

## Traversing a stack

````python
for i, item in enumerate(reversed(stack), 1):
    print(f"{-(i - len(stack) - 1)}) {item}")
````

````python
print("[")
print(*reversed(stack), sep="\n")
print("]")
````

## Removing Items from the Stack

````python
print(
    "Popped item:", stack.pop()
)  # pop() here both removes the last item, and returns it; the first item to be removed is "elderberry".
print("Remaining stack:", stack)
````

# Queue

Queue is a first-in-first-out (FIFO) data structure. *Deque* is used as the data structure "queue" in python.

Lists can also be turned into a queue.

Resource: [python docs](https://docs.python.org/3.13/tutorial/datastructures.html#using-lists-as-queues)

![Queue Image Display](queue.svg)

````python
from collections import deque
````

#### A Special Case: Priority Queue

In priority queues, the FIFO principle no longer applies, and elements are added and removed based on their *priority levels* to the queue.

Priority could be anything defined by the coder and is defined prior to the addition/removal tasks. 

An example could be seen below.

![Priority Queue Image Display](p_queue.svg)

## Insert an Element into Queue

````python
# Every element is added at the end of the queue.
queue = deque(["Patrick", "Christof", "Florin", "LÃ©onie"])
# In pseudo code, it would be "enqueue" to add elements into the queue.
queue.append("Michelle")
queue.append("Mykhailo")
queue
````

## Traversing a Queue

````python
for i, item in enumerate(queue, 1):
    print(f"{i}) {item}")
````

````python
print("<<")
print(*queue, sep="\n")
print(">>")
````

````python
import random

# Observing randomly who is in the queue along with their position.
rand_idx = random.randint(0, len(queue) - 1)
print(f"{rand_idx + 1}) {queue[rand_idx]}")
````

## Removing Items from the Queue

````python
# The element on the left side, at the front of the queue, is removed.
print(
    "Removed item:", queue.popleft()
)  # popleft() here both removes the first item, and returns it; the first item to be removed is "Patrick".

# In pseudo code, it would be "dequeue" to remove elements from the queue.

print("Remaining queue:", queue)
# Attention: queue.pop() would remove the element at the end of the queue which counteracts the typical behavior of a queue!
````
